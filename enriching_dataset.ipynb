```python
import pandas as pd
```

```python
from googleapiclient.discovery import build

api_key = "myApiKey"
youtube = build("youtube", "v3", developerKey=api_key)
```

```python
def fetch_video_details(video_ids):
    request = youtube.videos().list(
        part="kind, snippet, contentDetails, statistics",
        id=",".join(video_ids)  # Join multiple video IDs
    )
    response = request.execute()
    return response
```

```python
def save_to_csv(data, file_name="enriched_data.csv"):
    df = pd.DataFrame(data)
    df.to_csv(file_name, mode="a", header=False, index=False)
```

```python
video_IDs = []
with open("videoIds.txt", "r") as f:
    for line in f:
        video_IDs.append(line.strip())
```

```python
for i in range(5):
    print(video_IDs[i])

print(len(video_IDs))
```

```python
import time

mini_batch = 50
quota = 10000
for i in range(0, 10000, mini_batch):
    video_ids = video_IDs[i:i+mini_batch]
    try:
        response = fetch_video_details(video_ids)
        data = []
        for item in response.get("items", []):
            data.append({
            "video_id": item["id"],
            "title": item["snippet"]["title"],
            "description": item["snippet"]["description"],
            "published_at": item["snippet"]["publishedAt"],
            "tags": item["snippet"].get("tags", []),
            "category_id": item["snippet"]["categoryId"],
            "defaultAudioLanguage": item["snippet"].get("defaultAudioLanguage", "NA"),
            "duration": item["contentDetails"]["duration"],
            "channel": item["snippet"]["channelTitle"],
            "views": int(item["statistics"].get("viewCount", 0)),
            "likes": int(item["statistics"].get("likeCount", 0)),
            "comments": int(item["statistics"].get("commentCount", 0)),
            "engage": int(item["statistics"].get("likeCount", 0)) + int(item["statistics"].get("commentCount", 0)) / int(item["statistics"].get("viewCount", -1)),
            "caption": item["contentDetails"].get("caption", "NA"),
            "favorite": item["statistics"].get("favoriteCount", 0),
            })
            
    except Exception as e:
        print(f"Error with batch {i//mini_batch + 1}: {e}")
        continue

    save_to_csv(data)
    time.sleep(1)


    
```
