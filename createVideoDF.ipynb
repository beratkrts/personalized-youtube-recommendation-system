```python
import json

with open('Takeout-2/YouTube and YouTube Music/history/watch-history.json') as f:
    data = json.load(f)
```

```python
data = [d for d in data if 'titleUrl' in d]
```

```python
len(data)
```

```python
def extractVideoInfo(videoEntry):
    return {
        'title': videoEntry['title'],
        'url': videoEntry['titleUrl'],
        'time': videoEntry['time'],
        'channelName': videoEntry['subtitles'][0]['name'],
        'channelUrl': videoEntry['subtitles'][0]['url'],
        'videoId': videoEntry['titleUrl'].split('=')[-1].split('&')[0]
    }
```

```python
without_subtitles = [d for d in data if 'subtitles' not in d]

for d in without_subtitles:
    print(d)

print(len(without_subtitles))
```

```python
for d in data:
    if d in without_subtitles:
        d['subtitles'] = [{'name': 'Unknown', 'url': 'Unknown'}]    
```

```python
videos = [extractVideoInfo(d) for d in data]

import pandas as pd

df = pd.DataFrame(videos)
```

```python
df.head()
```

```python
df['title'] = df['title'].apply(lambda x: x[len('Watched '):])
```

```python
df.head()
```

```python
df.to_csv('youtube_watch_history.csv', index=False)
```

```python
batch_size = 1000  # Adjust based on your quota and system
for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i + batch_size]
    batch.to_csv(f"batch_{i//batch_size + 1}.csv", index=False)
```
